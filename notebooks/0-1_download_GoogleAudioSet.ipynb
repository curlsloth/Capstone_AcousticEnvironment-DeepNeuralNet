# Download the Google AudioSet

```python
import pandas as pd
```

```python
unbalanced_df = pd.read_csv('../audioset-processing/data/unbalanced_train_segments.csv',header=[0,1])
```

```python
unbalanced_df
```

```python
# get the youtube videos labeled as city/urban or rural/nature
urban_ind = unbalanced_df[unbalanced_df.columns[3]].str.contains('/t/dd00128')
natural_ind = unbalanced_df[unbalanced_df.columns[3]].str.contains('/t/dd00129')
```

```python
df_urban = unbalanced_df[urban_ind]
df_natural = unbalanced_df[natural_ind]
```

```python
# randomly sampled 10,000 files from more than 20,000 files
df_urban_sampled = df_urban.sample(n=10000, random_state=23, axis=0)
df_urban_sampled
```

```python
df_natural_sampled = df_natural.sample(n=10000, random_state=23, axis=0)
df_natural_sampled
```

```python
df = pd.concat([df_urban_sampled, df_natural_sampled])
df
```

```python
df.to_csv('../audioset-processing/data/unbalanced_train_segments_sampled10000_UrbanNatural.csv', index=False)
```
